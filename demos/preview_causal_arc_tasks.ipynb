# View CausalARC tasks

```python
import json
import platform
import os

# Custom modules.
os.chdir("../causal_arc")
from carc_utils import UtilsARC
from carc_augment import AugmentARC
from carc import CausalARC
from carc_tasks_counting import TaskCounting

# View versioning.
print("python version :", platform.python_version())
```

```python
u = UtilsARC()
a = AugmentARC()
c = CausalARC()
path = "../data/static_evaluation_set/v0_09-01-25/"
```

# Counting

```python
counting_task_file = "counting/causal_arc_counting.json"
with open(path+counting_task_file) as json_file:
    counting_task_dict = json.load(json_file)

counting_solution_file = "counting/causal_arc_counting_solutions.json"
with open(path+counting_solution_file) as json_file:
    counting_solution_dict = json.load(json_file)

uc = UtilsARC(task_dict = counting_task_dict,
              solution_dict = counting_solution_dict,
              submission_dict = None)
```

```python
for task in counting_task_dict.keys():
    uc.plot_demonstrations(task)
```

```python
# View test cases.
for task,tdict in counting_task_dict.items():
    print(task)
    test_input = tdict["test"][0]["input"]
    test_output = tdict["test"][0]["output"]
    u.plot_input_output(test_input, test_output, grid = False)
```
