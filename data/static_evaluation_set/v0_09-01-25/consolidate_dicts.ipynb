```python
# Load JSON files.
counting_file = "counting/causal_arc_counting.json"
with open(counting_file) as json_file:
    counting_dict = json.load(json_file)

extension_file = "extension/causal_arc_extension.json"
with open(extension_file) as json_file:
    extension_dict = json.load(json_file)

ordering_file = "ordering/causal_arc_ordering.json"
with open(ordering_file) as json_file:
    ordering_dict = json.load(json_file)

# Logical tasks.
path = "logical/"
logical_files = [f for f in listdir(path) if isfile(join(path, f))]
print(logical_files)
logical_dicts = []
for f in logical_files:
    if "solutions" not in f:
        with open(path+f) as json_file:
            logical_dicts.append(json.load(json_file))
```

# Counting

```python
for task,task_dict in counting_dict.items():
    scm_name = task.split("_")[0]
    scm = task_dict["scm"].replace("get_xy", scm_name)
    scm = scm.replace("\n    def", "def")
    scm = scm.replace("(colors):", "(colors: list):")
    task_dict["scm"] = scm
    print(scm)
    print()
```

# Extension

```python
for task,task_dict in extension_dict.items():
    scm_name = task.split("!")[0]
    scm = task_dict["scm"].replace("get_xy", scm_name)
    scm = scm.replace("\n        def", "def")
    scm = scm.replace("(colors):", "(colors: list):")
    task_dict["scm"] = scm
    print(scm)
    print()
```

# Ordering

```python
for task,task_dict in ordering_dict.items():
    scm_name = task.split("!")[0]
    scm = task_dict["scm"].replace("get_xy", scm_name)
    scm = scm.replace("\n        def", "def")
    scm = scm.replace("(colors):", "(colors: list):")
    task_dict["scm"] = scm
    print(scm)
    print()
```

# Logical

```python
# Consolidate.
all_logical_dict = dict()

# Rename tasks.
for d in logical_dicts:
    for task,task_dict in d.items():
        if "SCMdky5" in task:
            task_name = task.replace("SCMdky5", "SCMdky5!")
        else:
            task_name = task
        all_logical_dict[task_name] = task_dict
        print(task_name)
        print(task_dict["scm"])
```

```python
# Update SCMs.
for task,task_dict in all_logical_dict.items():
    scm_name = task.split("!")[0]
    print(scm_name)
    scm = task_dict["scm"].replace("get_xy", scm_name)
    scm = scm.replace("\n        def", "def")
    scm = scm.replace("(colors):", "(colors: list):")
    task_dict["scm"] = scm
    print(scm)
    print()
```
